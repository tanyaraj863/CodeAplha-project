```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://www.geeksforgeeks.org/python/python-programming-language-tutorial/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
print(soup.title.text)


headings = soup.find_all(["h1", "h2", "h3"])


data = []

for h in headings:
    title = h.text.strip()     
    tag_name = h.name         
    data.append([title, tag_name])


with open("gfg_tutorial_headings.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Heading", "Tag Type (h1/h2/h3)"])
    writer.writerows(data)

print("Finalized") 
```
