```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\HP\\Downloads\\sales_data_sample.csv", encoding="latin1")

print("FIRST 5 ROWS:")
print(df.head())

print("\nDATASET INFO:")
print(df.info())

print("\nSUMMARY STATISTICS:")
print(df.describe())

print("\nMISSING VALUES:")
print(df.isnull().sum())


print("\nTOTAL SALES BY PRODUCT LINE:")
print(df.groupby("PRODUCTLINE")["SALES"].sum())

print("\nAVERAGE SALES BY COUNTRY:")
print(df.groupby("COUNTRY")["SALES"].mean())

print("\nSALES COUNT PER MONTH:")
print(df["MONTH_ID"].value_counts().sort_index())

df.groupby("PRODUCTLINE")["SALES"].sum().plot(kind="bar")
plt.title("Total Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Sales")
plt.show()

df.groupby("COUNTRY")["SALES"].sum().sort_values().tail(10).plot(kind="bar")
plt.title("Sales by Country (Top 10)")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.show()

df.groupby("MONTH_ID")["SALES"].sum().plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month ID")
plt.ylabel("Total Sales")
plt.show()
```
